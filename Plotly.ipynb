```python
# Grafico de Barras
import plotly.express as px
df = px.data.gapminder().query("year == 2007")
fig = px.bar(df,x="continent",y="pop",color="country")
fig.show()
```

```python
# Grafico de pastel
import plotly.express as px
df = px.data.gapminder().query("year == 2007")
fig = px.pie(df,names="continent",title="Continente Más Popular")
fig.show()
```

```python
#Grafico de Lineas
import pandas as pd
import plotly.express as px
#Resumir grafico Ventas por dia
df = pd.DataFrame({'Dia':['Lunes','Martes','Miercoles','Jueves','Viernes'],
                   'Ventas':[100,120,90,80,110]})
fig = px.line(df,x='Dia',y='Ventas',title='Grafico de Lineas')
fig.show()
```

```python
#Grafico de Burbujas
import plotly.express as px
df = px.data.gapminder().query("year==2007")
fig = px.scatter(df,x="gdpPercap",y="lifeExp",size="pop",color="continent",hover_name="country",log_x=True,size_max=60)
fig.show()
```

```python
#Histograma
import plotly.express as px
import pandas as pd
df = pd.DataFrame({'Valores':[10,15,7,12,8,5,9,11,14,13]})
fig = px.histogram(df,x='Valores', title="Histograma")
fig.show()
```
